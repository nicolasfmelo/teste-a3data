```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

from unidecode import unidecode
```

```python
df_aircraft = pd.read_csv('aeronave.csv', sep=',')
df_ocurrence = pd.read_csv('ocorrencia.csv', sep=',')
```

```python
df_aircraft.shape
```

```python
df_ocurrence.shape
```

```python
##merge de colunas de aeronave e ocorrencia, facilita a manipulação do dataframe
df_ocurrence["tipo_motor"] = df_ocurrence["codigo_ocorrencia"].apply(lambda x: df_aircraft[df_aircraft["codigo_ocorrencia"]==x]["tipo_motor"].values[0])
df_ocurrence["equipamento"] = df_ocurrence["codigo_ocorrencia"].apply(lambda x: df_aircraft[df_aircraft["codigo_ocorrencia"]==x]["equipamento"].values[0])
df_ocurrence["quantidade_motores"] = df_ocurrence["codigo_ocorrencia"].apply(lambda x: df_aircraft[df_aircraft["codigo_ocorrencia"]==x]["quantidade_motores"].values[0])
df_ocurrence["nivel_dano"] = df_ocurrence["codigo_ocorrencia"].apply(lambda x: df_aircraft[df_aircraft["codigo_ocorrencia"]==x]["nivel_dano"].values[0])
df_ocurrence["categoria_aviacao"] = df_ocurrence["codigo_ocorrencia"].apply(lambda x: df_aircraft[df_aircraft["codigo_ocorrencia"]==x]["categoria_aviacao"].values[0])
df_ocurrence["fase_operacao"] = df_ocurrence["codigo_ocorrencia"].apply(lambda x: df_aircraft[df_aircraft["codigo_ocorrencia"]==x]["fase_operacao"].values[0])
```

```python
df_ocurrence.to_csv("./csv/composed.csv", sep=",")
```
